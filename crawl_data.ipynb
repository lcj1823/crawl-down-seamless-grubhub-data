```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from lxml import etree
import time
import csv
import pandas as pd

```

```python
url='https://www.seamless.com/menu/the-flame-diner-893-9th-ave-new-york/275128/reviews'

browser = webdriver.Chrome()
browser.get(url)

for i in range(40):
    html = etree.HTML(browser.page_source)
    x = html.xpath('//button[contains(text(),"See more")]')#whether there is 'see more'
    if x:#if there is see more
        browser.find_element_by_xpath('//button[contains(text(),"See more")]').click()#click
    else:
        pass
    time.sleep(3)
html = etree.HTML(browser.page_source)
browser.quit()#quit browser
reviewer = html.xpath('//span[@class="review-reviewer-name h6 u-margin-bottom-cancel"]/text()')#reviewer
comment = html.xpath('//p[@class="body u-stack-x-3 u-stack-y-3"]/text()')#comment
Time = html.xpath('//span[@class="meta-label"]/text()')#time
Label = html.xpath('//div[@class="u-inset-4 u-line-bottom u-line--thin u-line--light"]')#label


```
